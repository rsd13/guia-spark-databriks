# Introduccin
## Ejecucin distribuida de Apache Spark

Spark es un motor de procesamiento de datos distribuido con sus componentes trabajando en colaboracin en un clster de mquinas. 

Una aplicacin Spark consiste en un programa driver que es responsable de orquestar las operaciones operaciones paralelas en el clster de Spark. El driver accede a los componentes distribuidos en el clster, los **executers** de Spark y el **cluster manager** a travs de una SparkSession



### Spark driver

Como parte de Spark Aplication responsable de instanciar una SparkSession, el driver de Spark tiene mltiples funciones:
- se comunica con el Cluster Managerr
- solicita recursos (CPU, memoria, etc.) el Cluster Manager para los executers de Spark (JVMs)
- transforma todas las operaciones de Spark en clculos DAG, los programa y distribuye su ejecucin como taks entre los executors de Spark. Una vez asignados los recursos, se comunica directamente con los executors.

### SparkSession
En Spark 2.0, SparkSession comenz a ser un conductor unificado para todas las operaciones de Spark. Anteriormente para Spark tenias los siguientes puntos dee entrada: SparkContext, SQLContext, HiveContext, SparkConf, y StreamingContext. Ahora se unifica todo en SparkSession

### Cluster Manager
Es el responsable de administrar y asignar los recursos del cluster a los nodos que es donde se ejecuta las aplicaciones de Spark. Actualmente Spark soporta 4 cluster managers, Apache Hadoop, YARN, Apache Mesos y Kubernetes
### Spark executor
Spark executor se ejecuta en cada nodo trabajador del clster. Los ejecutores se comunican con el programa driver y son responsables de ejecutar las task en los trabajadores. En la mayora de los modos de despliegue, solo se ejecuta un nico ejecutor por nodo.


## Datos distribuidos y particiones

Los datos fsicos reales se distribuyen en el almacenamiento como particiones que residen en HDFS o en el almacenamiento en la nube. Mientras que los datos se distribuyen como particiones en el clster , Spark trata cada particin como una abstraccin de datos lgica de alto nivel, como un DataFrame en la memoria. 

Adems, a cada ejecutor de Spark se le asigna preferentemente una tarea que requiere que lea la particin ms cercana a l en la red, minimizando el ancho de banda de la red. Es decir, a cada ncleo del ejecutor se le asigna su propia particin de datos para trabajar observando la localidad de los datos
