# Introducción
## Ejecución distribuida de Apache Spark

Spark es un motor de procesamiento de datos distribuido con sus componentes trabajando en colaboración en un clúster de máquinas. 

Una aplicación Spark consiste en un programa driver que es responsable de orquestar las operaciones operaciones paralelas en el clúster de Spark. El driver accede a los componentes distribuidos en el clúster, los **executers** de Spark y el **cluster manager** a través de una SparkSession



### Spark driver

Como parte de Spark Aplication responsable de instanciar una SparkSession, el driver de Spark tiene múltiples funciones:
- se comunica con el Cluster Managerr
- solicita recursos (CPU, memoria, etc.) el Cluster Manager para los executers de Spark (JVMs)
- transforma todas las operaciones de Spark en cálculos DAG, los programa y distribuye su ejecución como taks entre los executors de Spark. Una vez asignados los recursos, se comunica directamente con los executors.

### SparkSession
En Spark 2.0, SparkSession comenzó a ser un conductor unificado para todas las operaciones de Spark. Anteriormente para Spark tenias los siguientes puntos dee entrada: SparkContext, SQLContext, HiveContext, SparkConf, y StreamingContext. Ahora se unifica todo en SparkSession

### Cluster Manager
Es el responsable de administrar y asignar los recursos del cluster a los nodos que es donde se ejecuta las aplicaciones de Spark. Actualmente Spark soporta 4 cluster managers, Apache Hadoop, YARN, Apache Mesos y Kubernetes
### Spark executor
Spark executor se ejecuta en cada nodo trabajador del clúster. Los ejecutores se comunican con el programa driver y son responsables de ejecutar las task en los trabajadores. En la mayoría de los modos de despliegue, solo se ejecuta un único ejecutor por nodo.


## Datos distribuidos y particiones

Los datos físicos reales se distribuyen en el almacenamiento como particiones que residen en HDFS o en el almacenamiento en la nube. Mientras que los datos se distribuyen como particiones en el clúster , Spark trata cada partición como una abstracción de datos lógica de alto nivel, como un DataFrame en la memoria. 

Además, a cada ejecutor de Spark se le asigna preferentemente una tarea que requiere que lea la partición más cercana a él en la red, minimizando el ancho de banda de la red. Es decir, a cada núcleo del ejecutor se le asigna su propia partición de datos para trabajar observando la localidad de los datos
