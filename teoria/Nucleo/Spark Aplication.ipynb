## Spark Application Concepts

- **Aplications**: Un programa de usuario construido sobre Spark utilizando sus APIs. Consiste en un programa controlador y ejecutores en el clúster
- **SparkSession**: Un objeto que proporciona un punto de entrada para interactuar con la funcionalidad subyacente de Spark y permite programar Spark con sus API. En un Spark Shell interactivo, el controlador de Spark instala una SparkSession automáticamente, mientras que en una aplicación de Spark, hay que crear un objeto SparkSession
- **Job**: un job que se ejecuta a través de cluster con una computación paralela que consiste en múltiples tareas que se generan en respuesta a una acción de Spark (por ejemplo, save(), collect())
- **Stage**: Cada job se divide en conjuntos de tareas más pequeñas, llamados stage(etapas), que dependen unos de otros.
- **Task**: Una única unidad de trabajo o ejecución que se enviará a un ejecutor de Spark

### Spark Application and SparkSession

En el núcleo de toda aplicación Spark está el programa diver de Spark, que crea un objeto SparkSession. Cuando trabajas con un shell de Spark, el driver es parte del shell y el objeto SparkSession (accesible a través de la variable spark) se crea para ti.

Si se lanza una la shell de Spark localmente en tu portátil, todas las operaciones se ejecutan localmente, en una única JVM. Pero puedes lanzar una shell de Spark para analizar datos en paralelo en un clúster con la misma facilidad que en modo local. Los comandos spark-shell --help o pyspark --help le mostrarán cómo conectarse al gestor de clusters de Spark

Una vez que tengas una SparkSession, puedes programar Spark usando las APIs para realizar operaciones de Spark.

### Spark Jobs

Cuando se ejecuta Spark, el Driver convierte su aplicación Spark en uno o más Spark Jobs. A continuación, transforma cada trabajo en un DAG. Esto, en esencia, es el plan de ejecución de Spark, donde cada nodo dentro de un DAG puede ser una o varias stages de Spark.





### Spark Stage

Como parte de los nodos DAG, se crean stages en función de las operaciones que pueden realizarse en serie o en paralelo ). No todas las operaciones de Spark pueden ocurrir en una sola stage, por lo que pueden dividirse en múltiples stage. 



### Spark Task

Cada stage se compone de task de Spark (una unidad de ejecución), que luego se relacionan en cada ejecutor de Spark; cada tarea se asigna a un solo núcleo y trabaja en una sola partición de datos (Figura 2-5). Así, un ejecutor con 16 núcleos puede tener 16 o más tareas trabajando en 16 o más particiones en paralelo, lo que hace que la ejecución de las tareas de Spark sea extremadamente paralela.
